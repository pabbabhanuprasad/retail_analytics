```python
import pandas as pd
import mysql.connector

# Connect to MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Bhanu@123",
    database="retail"
)

query = "SELECT * FROM transaction_products"
transaction_products = pd.read_sql(query, conn)

# Close connection
conn.close()

display(transaction_products.head(5))

transaction_products.to_csv("transaction_products1.csv", index=False)
```

```python
basket = transaction_products.groupby(['transaction_id','product_name'])['product_name'] \
        .count().unstack().fillna(0)

basket = (basket > 0).astype(int)

print(basket.head())
```

```python
import warnings
warnings.filterwarnings('ignore')
!pip install mlxtend -qq

```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(
    basket,
    min_support=0.02,
    use_colnames=True
)

print(frequent_itemsets.head(20))
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(
    frequent_itemsets,
    metric="lift",
    min_threshold=1
)

rules = rules[['antecedents','consequents','support','confidence','lift']]

display(rules.head())
```

```python
rules['antecedents'] = rules['antecedents'].apply(lambda x: ', '.join(list(x)))
rules['consequents'] = rules['consequents'].apply(lambda x: ', '.join(list(x)))

display(rules.head())
```

```python
rules.sort_values(by='lift', ascending=False).head(10)
```

```python
import pandas as pd
import mysql.connector

# Connect to MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Bhanu@123",
    database="retail"
)

# Load RFM base table
query = "SELECT * FROM rfm_base_new"
rfm1 = pd.read_sql(query, conn)

# Close connection
conn.close()

print("Data Loaded Successfully")
display(rfm1.head())


# -----------------------------
# RFM SCORING (1-5 SCALE)
# -----------------------------

# Recency score (lower recency = better)
rfm1['R_score'] = pd.qcut(rfm1['recency'], 5, labels=[5,4,3,2,1])

# Frequency score (higher frequency = better)
rfm1['F_score'] = pd.qcut(rfm1['frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])

# Monetary score (higher spending = better)
rfm1['M_score'] = pd.qcut(rfm1['monetary'], 5, labels=[1,2,3,4,5])


# Convert scores to integer
rfm1['R_score'] = rfm1['R_score'].astype(int)
rfm1['F_score'] = rfm1['F_score'].astype(int)
rfm1['M_score'] = rfm1['M_score'].astype(int)


# Create RFM Score
rfm1['RFM_score'] = (
    rfm1['R_score'].astype(str) + rfm1['F_score'].astype(str) + rfm1['M_score'].astype(str)
)

display("rfm_score",rfm1.head())



#customer segmentation
def segment_customer(row):

    if row['R_score'] >= 4 and row['F_score'] >= 4:
        return "Champions"

    elif row['R_score'] >= 3 and row['F_score'] >= 3:
        return "Loyal Customers"

    elif row['R_score'] <= 2 and row['F_score'] >= 3:
        return "At Risk"

    else:
        return "Hibernating"


rfm1['Segment'] = rfm1.apply(segment_customer, axis=1)

# Show segment distribution
print("\nCustomer Segments Distribution:")
display(rfm1['Segment'].value_counts())


rfm1.to_csv("rfm_customer_segments_new.csv", index=False)

print("\nRFM Segmentation Completed Successfully")
print("File saved as: rfm_customer_segments_new.csv")

# df5=pd.read_csv(r"C:\Users\Bhanu prasad\Downloads\rfm_customer_segments.csv")
# display(df5)
```
